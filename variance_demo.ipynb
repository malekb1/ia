# Variance, Standard Variance, and Écart-Type (Standard Deviation)

## Definitions
**Variance** measures how far data points are spread out from the mean.
$$
 \mathrm{Var}(X) = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2
$$

**Standard Deviation (Écart-Type)** is the square root of the variance.
$$
 \sigma = \sqrt{\mathrm{Var}(X)}
$$

```python
import numpy as np

data = np.array([2, 4, 4, 6, 8])

mean = data.mean()

variance = np.var(data)

std_dev = np.std(data)

print(f"Mean: {mean}, Variance: {variance}, Standard Deviation: {std_dev}")
```
